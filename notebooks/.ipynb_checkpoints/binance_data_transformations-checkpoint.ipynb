## Preparing sentiment

>What I need to do with the sentiment dataframe now is have each row be a single day and a column for Pos, Neg, and Neut that are just the means of every post from that day.

```python
rbinance_all.drop(columns='Unnamed: 0', inplace=True)
```

### dealing with dates

```python
rbinance_all.shape

rbinance_all['Date'] = pd.to_datetime(rbinance_all['easy_time'])


rbinance_all['DDate'] = rbinance_all['Date'].dt.date


rbinance_all['Date'] = rbinance_all['DDate']
rbinance_all = rbinance_all.drop(columns={'DDate', 'easy_time'})

reddit_notext = rbinance_all.drop(columns={'all_text'})

too_much = reddit_notext.groupby(['Date']).mean()

model_df = too_much
```
